## 作業 023：繪製 3D 函數

練習繪製 3D 函數：$z = x^2 - y^2$，$x$ 範圍於 $-5 \sim 5$，$y$ 範圍於 $-5 \sim 5$ 之間。（請完成以下程式碼）

```julia
using Plots
```

```julia
n = 1000
x = rand(-5:0.05:5,n)
y = rand(-5:0.05:5,n)
```

```julia
f(x,y) = x^2-y^2
```

```julia
p = plot(x,y,f,st=:surface,size=(1800,750))
```
